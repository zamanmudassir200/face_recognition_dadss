```python
import face_recognition
import cv2
import os
import json
import sys
```

```python
def recognize_face(input_image_path):
    """
    Simple face recognition script
    """
    # Load known faces
    known_faces = []
    known_names = []
    
    # Load the faces from the known_faces folder
    known_dir = "known_faces/"
    for filename in os.listdir(known_dir):
        if filename.endswith(('.jpg', '.png', '.jpeg')):
            image = face_recognition.load_image_file(os.path.join(known_dir, filename))
            encoding = face_recognition.face_encodings(image)[0]
            known_faces.append(encoding)
            known_names.append(filename.split('.')[0])
    
    # Load the input image
    input_image = face_recognition.load_image_file('man.jpg')
    input_encoding = face_recognition.face_encodings(input_image)
    
    if len(input_encoding) == 0:
        return {"status": "no_face_detected"}
    
    # Compare faces
    results = face_recognition.compare_faces(known_faces, input_encoding[0])
    
    if True in results:
        matched_index = results.index(True)
        return {
            "status": "matched",
            "person": known_names[matched_index],
            "confidence": "high"
        }
    else:
        return {"status": "not_matched"}

# For the use of command line
if __name__ == "__main__":
    if len(sys.argv) > 1:
        result = recognize_face(sys.argv[1])
        print(json.dumps(result))
    else:
        print('Usage: python face_recognition_script.py <image_path>')
```
